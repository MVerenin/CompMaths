```python
import matplotlib.pyplot as plt
import numpy as np
import math as m

arrX = np.linspace(0,5,100)
arrY = arrX*m.e**(-arrX**2)
plt.plot(arrX, arrY)
plt.show()
```

```python
xm = 2**(-0.5)
f = 0.5*xm*m.e**(-xm**2)
x1 = 0.3
x = 1.
while abs(x-x1)>0.0001:
    x = x1
    x1 = f*m.e**(x1**2)
x2=1.5
x=1.
while abs(x-x2)>0.0001:
    x=x2
    x2=(-m.log(f/x2))**0.5
f1 = x1*m.e**(-x1**2)
f2 = x2*m.e**(-x2**2)
plt.plot([x1, x2],[f1,f2],color ='black')
plt.plot(arrX, arrY, color='yellow')
plt.show()
print(x2-x1)
```
